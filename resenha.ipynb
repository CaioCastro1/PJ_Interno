```python
import qrcode
from PIL import Image

# Define the data you want to encode
data = "https://docs.google.com/forms/d/e/1FAIpQLSciKzOw_RGeY96hdgW8pg_k-ArCfgUZhvJ3NZAgQu5oWkxcLg/viewform"  # Replace with your desired URL or text

# Create QR code instance
qr = qrcode.QRCode(
    version=1,  # Controls the size of the QR Code
    error_correction=qrcode.constants.ERROR_CORRECT_H,  # Use high error correction to allow space for the logo
    box_size=10,  # Size of each box (pixels)
    border=4,  # Thickness of the border
)

# Add data to the QR code
qr.add_data(data)
qr.make(fit=True)

# Create the QR code image with custom colors
qr_img = qr.make_image(fill_color="#8C52FF").convert('RGB')

# Load the logo image
logo = Image.open("logo.png")  # Replace with your logo image path

if logo.mode in ('RGBA', 'LA') or (logo.mode == 'P' and 'transparency' in logo.info):
    white_bg_logo = Image.new("RGB", logo.size, (255, 255, 255))  # Create a white background
    white_bg_logo.paste(logo, (0, 0), logo)  # Paste the logo with transparency onto the white background
    logo = white_bg_logo

# Calculate the size of the logo relative to the QR code
qr_size = qr_img.size[0]  # Width of the QR code
logo_size = int(qr_size / 4)  # Set the logo size to be 1/4th of the QR code size
logo = logo.resize((logo_size, logo_size), Image.Resampling.LANCZOS)

# Calculate the position to place the logo (centered)
logo_position = ((qr_img.size[0] - logo.size[0]) // 2, (qr_img.size[1] - logo.size[1]) // 2)

# Paste the logo onto the QR code
qr_img.paste(logo, logo_position)

# Save the final QR code with the logo
qr_img.save("custom_qr_code_with_logo.png")
```

```python
import qrcode
from PIL import Image, ImageDraw

# Custom gradient function
def apply_gradient(img, start_color, end_color):
    width, height = img.size
    gradient = Image.new("RGB", img.size, start_color)
    draw = ImageDraw.Draw(gradient)

    for i in range(height):
        r = int(start_color[0] + (end_color[0] - start_color[0]) * (i / height))
        g = int(start_color[1] + (end_color[1] - start_color[1]) * (i / height))
        b = int(start_color[2] + (end_color[2] - start_color[2]) * (i / height))
        draw.line((0, i, width, i), fill=(r, g, b))

    return Image.blend(img, gradient, 0.5)

# Define the data you want to encode
data = "https://example.com"  # Replace with your desired URL or text

# Create QR code instance
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4,
)

# Add data to the QR code
qr.add_data(data)
qr.make(fit=True)

# Create the QR code image
qr_img = qr.make_image(fill_color="#000000", back_color="#FFFFFF").convert('RGB')

# Apply gradient to the QR code
start_color = (79, 27, 106)  # Dark purple
end_color = (140, 82, 255)  # Light purple
qr_img = apply_gradient(qr_img, start_color, end_color)

# Customize the QR code's "eyes" (the position markers)
draw = ImageDraw.Draw(qr_img)

# Set colors for the eyes
eye_color_outer = (140, 82, 255)  # Light purple
eye_color_inner = (255, 255, 255)  # White

# Get the size of the QR code box
box_size = 10
eye_size_outer = box_size * 7
eye_size_inner = box_size * 3

# Top-left corner
draw.rectangle([box_size, box_size, box_size + eye_size_outer, box_size + eye_size_outer], fill=eye_color_outer)
draw.rectangle([box_size * 3, box_size * 3, box_size * 6, box_size * 6], fill=eye_color_inner)

# Top-right corner
qr_width = qr_img.size[0]
draw.rectangle([qr_width - box_size * 8, box_size, qr_width - box_size, box_size + eye_size_outer], fill=eye_color_outer)
draw.rectangle([qr_width - box_size * 6, box_size * 3, qr_width - box_size * 3, box_size * 6], fill=eye_color_inner)

# Bottom-left corner
draw.rectangle([box_size, qr_img.size[1] - box_size * 8, box_size + eye_size_outer, qr_img.size[1] - box_size], fill=eye_color_outer)
draw.rectangle([box_size * 3, qr_img.size[1] - box_size * 6, box_size * 6, qr_img.size[1] - box_size * 3], fill=eye_color_inner)

# Save the result as a PNG
qr_img.save("cool_qr_code_with_gradient.png")
```

```python
from moviepy.editor import VideoFileClip

def reduce_video_size(file_path, output_path, resolution=(640, 360), fps=24):
    # Load the video
    video = VideoFileClip(file_path)
    
    # Resize the video
    video_resized = video.resize(newsize=resolution)
    
    # Write the resized video with the desired frame rate
    video_resized.write_videofile(output_path, fps=fps, codec='libx264', preset='slow', bitrate='500k')
    
    print(f"Video saved with reduced resolution and frame rate at {output_path}")

# Usage example
reduce_video_size('/Users/caiocastro/Padrao/GitHub/PJ_Interno/Screen Recording 2024-10-22 at 20.44.11.mov', 'compressed_video.mp4', resolution=(640, 360), fps=24)
```
