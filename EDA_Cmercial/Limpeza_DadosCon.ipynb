```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, shapiro, levene, mannwhitneyu
from scipy.stats import chi2_contingency
from fuzzywuzzy import process

# remov column limit
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('bases/PAC.csv')
```

```python
df
```

```python
# Write a regex expression to capture the date bewteen the parenthesis in "Semana"

df['Semana'] = df['Semana'].str.extract(r'\((\d{2}/\d{2}/\d{4})\)')

df
```

```python
# Drop the first 20000 rows

df = df.iloc[20000:]

```

```python
df
```

```python
# Convert the "Semana" column to datetime

df['Semana'] = pd.to_datetime(df['Semana'])
```

```python
df
```

```python
# Only get the rows where the "Semana" column is greater than 2023-01-01

df = df[df['Semana'] > '2023-01-01']
```

```python
df
```

```python
# drop the last 10 columns

df = df.iloc[:, :-10]
```

```python
# Count different portfolios

df['Portfólio'].value_counts()
```

```python
df.dropna(subset=['Prospecção'], inplace=True)

df['Prospecção'] = df['Prospecção'].apply(lambda x: process.extractOne(x, ['E-mail', 'LinkedIn'])[0])
```

```python
# download the df to a csv file

df.to_csv('bases/PAC_cleaned.csv', index=False)
```
