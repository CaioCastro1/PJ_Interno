```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, shapiro, levene, mannwhitneyu
from scipy.stats import chi2_contingency
from fuzzywuzzy import process

# remov column limit
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('bases/PAC_CIV.csv')
```

```python
df
```

```python
# Write a regex expression to capture the date bewteen the parenthesis in "Semana"

df['Semana'] = df['Semana'].str.extract(r'\((\d{2}/\d{2}/\d{4})\)')

df
```

```python
# Convert the "Semana" column to datetime. It is d/m/y format

# Function to convert mixed date formats
def parse_dates(date_str):
    for fmt in ('%d/%m/%Y', '%m/%d/%Y'):
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            continue
    return pd.NaT


df['Semana'] = df['Semana'].apply(parse_dates)
```

```python
df
```

```python
# Only get the rows where the "Semana" column is greater than 2023-01-01

df = df[df['Semana'] > '2023-01-01']
```

```python
df
```

```python
# Count different portfolios

df['Portfólio'].value_counts()
```

```python
df.dropna(subset=['Prospecção'], inplace=True)

df['Prospecção'] = df['Prospecção'].apply(lambda x: process.extractOne(x, ['E-mail', 'LinkedIn'])[0])
```

```python
# Drop the last 5 columns

df.drop(df.columns[-5:], axis=1, inplace=True)
```

```python
# download the df to a csv file

df.to_csv('bases/PAC_NTECcleaned.csv', index=False)
```
